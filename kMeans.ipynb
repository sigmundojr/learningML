## K-means algorithm
The main goal is to find clusters using the Iris dataset

```
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb

%matplotlib inline
```

```
df = pd.read_csv('iris.csv')
```

```
df.head()
```

```
# trying to identify classes
sb.pairplot(df)
```

```
# passing the class
sb.pairplot(df,hue='target')
```

```
# X = object features, y = classes 
# dropping the target column
X = np.array(df.drop('target', axis = 1))
```

```
# importing KMeans
from sklearn.cluster import KMeans
```

```
# defining the number of clusters and the random
kmeans = KMeans(n_clusters=3, random_state = 0)
```

```
# training Kmeans
kmeans.fit(X)
```

```
# seeing the lables
kmeans.labels_
```

```
# comparing classes 
# creating a new column into the same dataframe
df['k-classes'] = kmeans.labels_
df.head()
```

```
# Generating visualization using the kMeans
sb.pairplot(df, hue='k-classes')
```
